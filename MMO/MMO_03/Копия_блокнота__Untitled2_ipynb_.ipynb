```python
import pandas as pd
file_path = "diabetes.csv"
df = pd.read_csv(file_path)

# –í—ã–≤–æ–¥–∏–º –æ–±—â—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –¥–∞–Ω–Ω—ã—Ö
df.info(), df.head()
```

```python
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω—É–ª–µ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö
zero_counts = df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].eq(0).sum()
zero_counts
```

```python

# –ó–∞–º–µ–Ω–∞ –Ω—É–ª–µ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –Ω–∞ –º–µ–¥–∏–∞–Ω—É
def replace_zeros_with_median(column):
    median_value = column.median()
    return column.replace(0, median_value)

# –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å—Ç–æ–ª–±—Ü–æ–≤
for column in ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI']:
    df[column] = replace_zeros_with_median(df[column])

# –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å—Ç–æ–ª–±—Ü–∞ Insulin
# –í—ã–±–æ—Ä: —É–¥–∞–ª–∏—Ç—å –∏–ª–∏ –∑–∞–º–µ–Ω–∏—Ç—å
# df.drop(columns=['Insulin'], inplace=True)  # –£–¥–∞–ª–µ–Ω–∏–µ
df['Insulin'] = replace_zeros_with_median(df['Insulin'])  # –ó–∞–º–µ–Ω–∞ –Ω–∞ –º–µ–¥–∏–∞–Ω—É

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –Ω—É–ª–µ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏
zero_counts_after = df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].eq(0).sum()
print(zero_counts_after)

# –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö (–ø—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
normalized_columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
df[normalized_columns] = scaler.fit_transform(df[normalized_columns])

print(df.head())
```

```python
Y = df['Outcome']  # –í–µ–∫—Ç–æ—Ä –º–µ—Ç–æ–∫
X = df.drop(columns=['Outcome'])  # –ú–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —É–¥–∞–ª—è–µ–º —Å—Ç–æ–ª–±–µ—Ü 'Outcome'

print("–í–µ–∫—Ç–æ—Ä –º–µ—Ç–æ–∫ Y:")
print(Y.head())
print("\n–ú–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ X:")
print(X.head())
```

```python
from sklearn.model_selection import train_test_split
#   est_size=0.2 —É–∫–∞–∑—ã–≤–∞–µ—Ç, —á—Ç–æ 20% –¥–∞–Ω–Ω—ã—Ö –æ—Ç–≤–æ–¥–∏—Ç—Å—è –Ω–∞ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É.
#random_state=42 —Ñ–∏–∫—Å–∏—Ä—É–µ—Ç —Å–ª—É—á–∞–π–Ω–æ–µ —Ä–∞–∑–±–∏–µ–Ω–∏–µ –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏.
# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("–†–∞–∑–º–µ—Ä –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏ X_train:", X_train.shape)
print("–†–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ X_test:", X_test.shape)
print("–†–∞–∑–º–µ—Ä –≤–µ–∫—Ç–æ—Ä–∞ –º–µ—Ç–æ–∫ Y_train:", Y_train.shape)
print("–†–∞–∑–º–µ—Ä –≤–µ–∫—Ç–æ—Ä–∞ –º–µ—Ç–æ–∫ Y_test:", Y_test.shape)
```

```python
#DecisionTreeClassifier: –≠—Ç–æ –∫–ª–∞—Å—Å, –∫–æ—Ç–æ—Ä—ã–π —Ä–µ–∞–ª–∏–∑—É–µ—Ç –∞–ª–≥–æ—Ä–∏—Ç–º –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π –¥–ª—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏.
#fit: –ú–µ—Ç–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –æ–±—É—á–∞–µ—Ç –º–æ–¥–µ–ª—å –Ω–∞ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.
from sklearn.tree import DecisionTreeClassifier

# –°–æ–∑–¥–∞–Ω–∏–µ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ –º–æ–¥–µ–ª–∏ –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π
model = DecisionTreeClassifier(random_state=42)

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
model.fit(X_train, Y_train)

# –ü—Ä–æ–≤–µ—Ä–∫–∞, —á—Ç–æ –º–æ–¥–µ–ª—å –æ–±—É—á–µ–Ω–∞
print("–ú–æ–¥–µ–ª—å –æ–±—É—á–µ–Ω–∞.")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –º–µ—Ç–æ–∫ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
Y_pred = model.predict(X_test)

# –û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏
accuracy = accuracy_score(Y_test, Y_pred)
print(f"–¢–æ—á–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ (Accuracy): {accuracy:.2f}")

# –ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫
conf_matrix = confusion_matrix(Y_test, Y_pred)
print("\n–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫:")
print(conf_matrix)

# –û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
class_report = classification_report(Y_test, Y_pred)
print("\n–û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(class_report)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø–æ–¥–±–æ—Ä–∞
#max_depth: –≠—Ç–æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –≥–ª—É–±–∏–Ω—É –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π. –ì–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤–∞ ‚Äî —ç—Ç–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—Ä–æ–≤–Ω–µ–π –æ—Ç –∫–æ—Ä–Ω—è –¥–æ –ª–∏—Å—Ç–∞.
#max_features: –≠—Ç–æ—Ç –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä –∑–∞–¥–∞–µ—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—é—Ç—Å—è –ø—Ä–∏ –ø–æ–∏—Å–∫–µ –ª—É—á—à–µ–≥–æ —Ä–∞–∑–±–∏–µ–Ω–∏—è –≤ –∫–∞–∂–¥–æ–º —É–∑–ª–µ –¥–µ—Ä–µ–≤–∞.
param_grid = {
    'max_depth': [None, 5, 10, 15, 20],
    'max_features': ['sqrt', 'log2', None]  # –£–±—Ä–∞–Ω–æ 'auto'
}

# –°–æ–∑–¥–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏ –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π
model = DecisionTreeClassifier(random_state=42)

# –°–æ–∑–¥–∞–Ω–∏–µ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=param_grid,
                           scoring='accuracy', cv=5, n_jobs=-1, verbose=1)

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ —Å –ø–æ–¥–±–æ—Ä–æ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
grid_search.fit(X_train, Y_train)

# –í—ã–≤–æ–¥ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –ª—É—á—à–µ–π —Ç–æ—á–Ω–æ—Å—Ç–∏
print(f"–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã: {grid_search.best_params_}")
print(f"–õ—É—á—à–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å: {grid_search.best_score_:.2f}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –ª—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –Ω–æ–≤–æ–π –º–æ–¥–µ–ª–∏
best_model = DecisionTreeClassifier(max_depth=grid_search.best_params_['max_depth'],
                                     max_features=grid_search.best_params_['max_features'],
                                     random_state=42)

# –û–±—É—á–µ–Ω–∏–µ –Ω–æ–≤–æ–π –º–æ–¥–µ–ª–∏ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
best_model.fit(X_train, Y_train)

# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –º–µ—Ç–æ–∫ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
Y_pred = best_model.predict(X_test)

# –û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏
accuracy = accuracy_score(Y_test, Y_pred)
print(f"–¢–æ—á–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ –ø–æ—Å–ª–µ —É–ª—É—á—à–µ–Ω–∏—è: {accuracy:.2f}")

# –ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫
conf_matrix = confusion_matrix(Y_test, Y_pred)
print("\n–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫:")
print(conf_matrix)

# –û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
class_report = classification_report(Y_test, Y_pred)
print("\n–û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏:")
print(class_report)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# –°–æ–∑–¥–∞–Ω–∏–µ –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ KNN
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, Y_train)
```

```python
# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –º–µ—Ç–æ–∫ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
Y_pred_knn = knn_model.predict(X_test)

# –û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏
accuracy_knn = accuracy_score(Y_test, Y_pred_knn)
print(f"–¢–æ—á–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ KNN: {accuracy_knn:.2f}")

# –ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫
conf_matrix_knn = confusion_matrix(Y_test, Y_pred_knn)
print("\n–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫ KNN:")
print(conf_matrix_knn)

# –û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
class_report_knn = classification_report(Y_test, Y_pred_knn)
print("\n–û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ KNN:")
print(class_report_knn)
```

```python
from sklearn.model_selection import GridSearchCV

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø–æ–¥–±–æ—Ä–∞
#–ü–æ —É–º–æ–ª—á–∞–Ω–∏—é –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è 
#ùëò
#=
#5
#k=5 (5 –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π) –∏ weights='uniform' (–≤—Å–µ —Å–æ—Å–µ–¥–∏ –∏–º–µ—é—Ç —Ä–∞–≤–Ω—ã–π –≤–µ—Å).
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance']
}

# –°–æ–∑–¥–∞–Ω–∏–µ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ GridSearchCV –¥–ª—è KNN
grid_search_knn = GridSearchCV(estimator=knn_model, param_grid=param_grid_knn,
                                scoring='accuracy', cv=5, n_jobs=-1, verbose=1)

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ —Å –ø–æ–¥–±–æ—Ä–æ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
grid_search_knn.fit(X_train, Y_train)

# –í—ã–≤–æ–¥ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –ª—É—á—à–µ–π —Ç–æ—á–Ω–æ—Å—Ç–∏
print(f"–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã KNN: {grid_search_knn.best_params_}")
print(f"–õ—É—á—à–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å KNN: {grid_search_knn.best_score_:.2f}")
```

```python
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –ª—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –Ω–æ–≤–æ–π –º–æ–¥–µ–ª–∏ KNN
best_knn_model = KNeighborsClassifier(n_neighbors=grid_search_knn.best_params_['n_neighbors'],
                                      weights=grid_search_knn.best_params_['weights'])

# –û–±—É—á–µ–Ω–∏–µ –Ω–æ–≤–æ–π –º–æ–¥–µ–ª–∏ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
best_knn_model.fit(X_train, Y_train)

# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –º–µ—Ç–æ–∫ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
Y_pred_best_knn = best_knn_model.predict(X_test)

# –û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏
accuracy_best_knn = accuracy_score(Y_test, Y_pred_best_knn)
print(f"–¢–æ—á–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ KNN –ø–æ—Å–ª–µ —É–ª—É—á—à–µ–Ω–∏—è: {accuracy_best_knn:.2f}")

# –ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫
conf_matrix_best_knn = confusion_matrix(Y_test, Y_pred_best_knn)
print("\n–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫ KNN –ø–æ—Å–ª–µ —É–ª—É—á—à–µ–Ω–∏—è:")
print(conf_matrix_best_knn)

# –û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
class_report_best_knn = classification_report(Y_test, Y_pred_best_knn)
print("\n–û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ KNN –ø–æ—Å–ª–µ —É–ª—É—á—à–µ–Ω–∏—è:")
print(class_report_best_knn)
```

```python
from sklearn.ensemble import RandomForestClassifier

# –°–æ–∑–¥–∞–Ω–∏–µ –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, Y_train)
```

```python
# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –º–µ—Ç–æ–∫ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
Y_pred_rf = rf_model.predict(X_test)

# –û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏
accuracy_rf = accuracy_score(Y_test, Y_pred_rf)
print(f"–¢–æ—á–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞: {accuracy_rf:.2f}")

# –ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫
conf_matrix_rf = confusion_matrix(Y_test, Y_pred_rf)
print("\n–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞:")
print(conf_matrix_rf)

# –û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
class_report_rf = classification_report(Y_test, Y_pred_rf)
print("\n–û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞:")
print(class_report_rf)
```

```python
# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –ø–æ–¥–±–æ—Ä–∞
#n_estimators: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ—Ä–µ–≤—å–µ–≤ –≤ –ª–µ—Å—É ([50, 100, 200]).
#max_depth: –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –≥–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤—å–µ–≤ ([None, 10, 20], –≥–¥–µ None –æ–∑–Ω–∞—á–∞–µ—Ç –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è).
#max_features: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º—ã—Ö –ø—Ä–∏ –ø–æ–∏—Å–∫–µ –ª—É—á—à–µ–≥–æ —Ä–∞–∑–±–∏–µ–Ω–∏—è (['sqrt', 'log2', None], –≥–¥–µ None –æ–∑–Ω–∞—á–∞–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –≤—Å–µ—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤).
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'max_features': ['sqrt', 'log2', None]  # –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ 'auto' —É–±—Ä–∞–Ω–æ
}

# –°–æ–∑–¥–∞–Ω–∏–µ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ GridSearchCV –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
grid_search_rf = GridSearchCV(estimator=rf_model, param_grid=param_grid_rf,
                               scoring='accuracy', cv=5, n_jobs=-1, verbose=1)

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ —Å –ø–æ–¥–±–æ—Ä–æ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
grid_search_rf.fit(X_train, Y_train)

# –í—ã–≤–æ–¥ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –ª—É—á—à–µ–π —Ç–æ—á–Ω–æ—Å—Ç–∏
print(f"–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞: {grid_search_rf.best_params_}")
print(f"–õ—É—á—à–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞: {grid_search_rf.best_score_:.2f}")
```

```python
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –ª—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –Ω–æ–≤–æ–π –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞
best_rf_model = RandomForestClassifier(n_estimators=grid_search_rf.best_params_['n_estimators'],
                                       max_depth=grid_search_rf.best_params_['max_depth'],
                                       max_features=grid_search_rf.best_params_['max_features'],
                                       random_state=42)

# –û–±—É—á–µ–Ω–∏–µ –Ω–æ–≤–æ–π –º–æ–¥–µ–ª–∏ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
best_rf_model.fit(X_train, Y_train)

# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –º–µ—Ç–æ–∫ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
Y_pred_best_rf = best_rf_model.predict(X_test)

# –û—Ü–µ–Ω–∫–∞ —Ç–æ—á–Ω–æ—Å—Ç–∏
accuracy_best_rf = accuracy_score(Y_test, Y_pred_best_rf)
print(f"–¢–æ—á–Ω–æ—Å—Ç—å –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –ø–æ—Å–ª–µ —É–ª—É—á—à–µ–Ω–∏—è: {accuracy_best_rf:.2f}")

# –ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫
conf_matrix_best_rf = confusion_matrix(Y_test, Y_pred_best_rf)
print("\n–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –ø–æ—Å–ª–µ —É–ª—É—á—à–µ–Ω–∏—è:")
print(conf_matrix_best_rf)

# –û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
class_report_best_rf = classification_report(Y_test, Y_pred_best_rf)
print("\n–û—Ç—á–µ—Ç –æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ –ø–æ—Å–ª–µ —É–ª—É—á—à–µ–Ω–∏—è:")
print(class_report_best_rf)
```
